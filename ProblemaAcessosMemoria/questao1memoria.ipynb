# Tratando os dados .txt e transformando no DataFrame em Excel.

```python
import pandas as pd
import os 

pasta = r'C:\Users\clodo\OneDrive\Área de Trabalho\BenchMark Acesso Memoria'


dataframes = []

for arquivo in os.listdir(pasta):
    if arquivo.endswith('.txt'):
        caminho_arquivo = os.path.join(pasta, arquivo)
        print(f'Lendo arquivo: {caminho_arquivo}')

        df_temp = pd.read_csv(caminho_arquivo, delimiter='\t')

        df_temp['KBinMemory'] = df_temp['KBinMemory'].astype(int)
    
        df_ordenado = df_temp.sort_values(by='KBinMemory', ascending = True)   

        dataframes.append(df_ordenado)


df_final_p2_q1 = pd.concat(dataframes, ignore_index= True)
df_final_p2_q1 = df_final_p2_q1.sort_values(by='KBinMemory', ascending=True).reset_index(drop=True)

print(df_final_p2_q1.head())

df_final_p2_q1.to_excel('dados_consolidados_problema2_q1.xlsx', index=False)

print("Todos os arquivos foram processados e salvos em 'dados_consolidados_problema2_q1.xlsx'.")
```

# Apresentando DataFrame

```python
df_final_p2_q1.rename(columns= {'ArraySize' : 'TamanhoArray', 'SearchRepetitions' : 'Repeticoes', 'StartTime' : 'TempoInicial', 'PopulateTime' : 'TempoPreenchido', 'SearchTime' : 'TempoDeBusca', 'TotalTime' : 'TempoTotal'},inplace= True)
df_final_p2_q1
```

# Questão 1 - Alternativa A 
Como o tempo de busca no array varia com o aumento do tamanho do array?
* Apresentando o primeiro gráfico com todos os Arrays e outros 3 com Arrays separados e filtrados para uma melhor observação.

```python
import matplotlib.pyplot as plt

tamanho_array = df_final_p2_q1['TamanhoArray']  
tempo_busca = df_final_p2_q1['TempoDeBusca']
primeros_dados = df_final_p2_q1.iloc[:7]
segundos_dados = df_final_p2_q1.iloc[7:14]
terceiros_dados = df_final_p2_q1.iloc[14:]




plt.figure(figsize=(10, 6))
plt.plot(tamanho_array, tempo_busca, marker='o', linestyle='-', color='orange')
plt.title('Variação do Tempo de Busca no Array com o Aumento do Tamanho do Array', fontsize=14)
plt.xlabel('Tamanho do Array (KB)', fontsize=12)
plt.ylabel('Tempo de Busca (ms)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize= (10, 6))
plt.plot(primeros_dados['TamanhoArray'], primeros_dados['TempoDeBusca'], marker = 'o', linestyle = '-', color = 'red')
plt.title('Variação do Tempo de Busca (Primeiros 7 Arrays)', fontsize = 14)
plt.xlabel('Tamanho do Array (KB)', fontsize = 12)
plt.ylabel('Tempo de Busca (ms)', fontsize = 12)
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize= (10, 6))
plt.plot(segundos_dados['TamanhoArray'], segundos_dados['TempoDeBusca'], marker = 'o', linestyle = '-', color = 'blue')
plt.title('Variação do Tempo de Busca (Próximos 7 Arrays)', fontsize = 14)
plt.xlabel('Tamanho do Array (KB)', fontsize = 12)
plt.ylabel('Tempo de Busca (ms)', fontsize = 12)
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize= (10, 6))
plt.plot(terceiros_dados['TamanhoArray'], terceiros_dados['TempoDeBusca'], marker = 'o', linestyle = '-', color = 'green')
plt.title('Variação do Tempo de Busca (Arrays Restantes)', fontsize = 14)
plt.xlabel('Tamanho do Array (KB)', fontsize = 12)
plt.ylabel('Tempo de Busca (ms)', fontsize = 12)
plt.grid(True)
plt.tight_layout()
plt.show()

```

# Questão Extra - Problema 2
* Análise do Ponto de Saturação da Cache.

```python
import matplotlib.pyplot as plt
  
dados_filtrados = df_final_p2_q1[(df_final_p2_q1['TamanhoArray'] >= 32) & (df_final_p2_q1['TamanhoArray'] <= 65536)]  # 32 KB a 64 MB

# Criar o gráfico novamente com os dados filtrados
plt.figure(figsize=(12, 6))
plt.plot(dados_filtrados['TamanhoArray'], dados_filtrados['TempoDeBusca'], marker='o', linestyle='-', color='blue', label='Tempo de Busca')

# Destacar faixas típicas de cache
plt.axvspan(32, 128, color='yellow', alpha=0.3, label='Cache L1 (32 KB - 128 KB)')
plt.axvspan(256, 2048, color='orange', alpha=0.3, label='Cache L2 (256 KB - 2 MB)')
plt.axvspan(4096, 65536, color='red', alpha=0.2, label='Cache L3 (4 MB - 64 MB)')

# Títulos e legendas
plt.title('Ponto de Saturação do Cache (Dados Filtrados)', fontsize=14)
plt.xlabel('Tamanho do Array (KB)', fontsize=12)
plt.ylabel('Tempo de Busca (ms)', fontsize=12)
plt.legend()
plt.grid(True)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```
